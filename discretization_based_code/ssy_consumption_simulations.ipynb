```julia
using PyPlot
plt = PyPlot
```

```julia
ssy = SSYconsumption()
```

```julia
ssyd1 = discretize(ssy, 10, 10, 10)
```

```julia
sc1, sz1, z1, c1 = sim_consumption(ssyd1, ts_length=80000);
```

```julia
ssyd2 = discretize(ssy, 50, 10, 10)
```

```julia
sc2, sz2, z2, c2 = sim_consumption(ssyd2, ts_length=80000);
```

```julia
mean(c1)
```

```julia
std(c1)
```

```julia
mean(c2)
```

```julia
std(c2)
```

```julia
fig, ax = plt.subplots()
ax[:plot](c1, alpha=0.75)
ax[:plot](c2, "r-", alpha=0.5)
ax[:set](ylim=(-0.05, 0.05))
plt.show()
```

```julia
fig, ax = plt.subplots()
ax[:plot](z1, alpha=0.75)
ax[:plot](z2, "r-", alpha=0.5)
plt.show()
```
