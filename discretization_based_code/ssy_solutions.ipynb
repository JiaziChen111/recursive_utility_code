## Computing Solutions, SSY Model

```julia
include("src/compute_fixed_points.jl")
```

```julia
using PyPlot
plt = PyPlot
```

```julia
K, I, J = 16, 16, 16
```

```julia
ez = EpsteinZinSSY()
```

```julia
ssy = SSYconsumption()
ssyd = discretize(ssy, K, I, J)
```

```julia
w = compute_fp(ez, ssyd, init_val=1000)
```

Plot w as a function of $\sigma_c$ and $\sigma_z$

```julia
i = 16
W = Array{Float64}(K, J)
for k in 1:K
    for j in 1:J
        m = multi_to_single(k, i, j, I, J)
        W[k, j] = w[m]
    end
end
```

```julia
x = ssyd.σ_c_states
y = ssyd.z_states[i,:];
```

```julia
fig, ax = plt.subplots(figsize=(10, 5.7))

cs1 = ax[:contourf](x, 
                    y, 
                    W',
                    alpha=0.6)


plt.colorbar(cs1, ax=ax)
```
