## Impact of discretization, SSY model

```julia
include("src/stability_tests.jl")
```

```julia
using PyPlot
plt = PyPlot
```

```julia
ez = EpsteinZinBY()
ssy = SSYconsumption()
```

```julia
I = 5
J = 5
index_vals = [5, 10, 15, 20, 25, 30, 35, 40, 45]

r_vals = [compute_spec_rad(ez, discretize(ssy, ix, I, J)) for ix in index_vals]
```

```julia
fig, ax = plt.subplots(figsize=(7.2, 4.2))

ax[:plot](index_vals, r_vals)
ax[:set_ylabel]("spectral radius")
ax[:set_xlabel]("number of grid points")
ax[:grid]()
ax[:set_ylim](1.115, 1.175)
ax[:set_xlim](minimum(index_vals), maximum(index_vals))

ax[:legend](loc="lower right")

#plt.savefig("r_vs_j.pdf")
plt.show()
```
