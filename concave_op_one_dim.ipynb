## Stability properties: Insights from one dimension

We are interested in a map 

$$ Tv(x) = a + \left( \int v^\theta(y) k(x, y) dy \right)^{1/\theta} $$

where $a > 0$.  

When we study it using fixed point theory, we first apply the map $\Phi v = v^\theta$ to points (functions like $v$, $Tv$) in the space.  This is a simple contiuous transformation and does not change the stability properties.

Essentially, this means that we are studying the operator

$$ Ag(x) = \left( a + \left( \int g(y) k(x, y) dy \right)^{1/\theta} \right)^\theta $$

or 

$$ Ag(x) = \left( a + ( K g(x)  )^{1/\theta} \right)^\theta $$

where $K$ is the integral operator associated with $k$.

Let's see how this map $A$ looks in the one-dimensional case:

$$ A(t) = \left( a + ( b t  )^{1/\theta} \right)^\theta $$

where $b > 0$ and $t \geq 0$.

Note that, in this case, the value $b$ corresponds to the spectral radius of the linear operator $K$.


We start with $\theta < 0$ and then consider other cases.

### The case $\theta < 0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = -2
a = 1
```

```python
def A(t, a, b, theta):
    return (a + (b * t)**(1/theta))**theta
```

#### Sub-case 1: $b < 1$

```python
b = 0.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 1, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

It's a contraction but there's no nontrivial fixed point.

#### Sub-case 2: $b > 1$

```python
b = 2.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, .5, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

It's not a contraction but it's concave and increasing and maps some point up.  This is the stable case.

### The case $0 < \theta < 1$

```python
theta = 0.5
```

#### Sub-case 1: $b < 1$

```python
b = 0.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 2, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

It's a contraction and there's a nontrivial fixed point.

#### Sub-case 2: $b > 1$

```python
b = 2.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 1, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

```python
b = 1.0
a = 0.002
theta = 0.4

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 0.5, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

There's no fixed point, although I think there would be for smaller $a$.

### The case $1 < \theta$

```python
theta = 1.5
a = 1
```

#### Sub-case 1: $b < 1$

```python
b = 0.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 10, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

Same thing with larger $\theta$

```python

theta = 2

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 20, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

Note that it looks concave and it's definitely not a contraction.  But I can't prove that it's concave.  There's a fixed point because the function is positive at zero and the slope is eventually $b$.

#### Sub-case 2: $b > 1$

```python
b = 1.5

fig, ax = plt.subplots(figsize=(10, 8))
t_vals = np.linspace(0.001, 4, 200)

ax.plot(t_vals, t_vals, label="45 degrees")
ax.plot(t_vals, A(t_vals, a, b, theta), label="$S$")
ax.legend()
plt.show()
```

There's no fixed point because the slope is eventually $b > 1$ and the function is positive at zero.
