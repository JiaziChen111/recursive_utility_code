```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib import rc

rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
def ϕ(t, θ):
    return (1 + t**(1/θ))**θ
```

First let's see the function $Ag = \phi(Kg)$ when $g$ is scalar and $K$ is just a constant multiple, either greater or less than one:

```python

def plot_pairs(θ=1.0, grid_max=2.0):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    g = np.linspace(1e-8, grid_max, 200)

    titles = "$r(K) > 1$", "$r(K) \leq 1$"

    for (K, ax, title) in zip ((1.5, 0.5), axes, titles):

        ax.plot(g, g, 'k--', label="$45^\circ$")
        ax.plot(g, ϕ(K * g, θ), label="$\phi(Kg)$")
        ax.legend()
        ax.set_xlabel("$g$", fontsize=14)
        ax.set_ylabel("$Ag$", fontsize=14)
        ax.set_ylim(0, grid_max)
        ax.set_xlim(0, grid_max)

        ax.set_title(title)

    plt.show()
```

We'll look at $\theta < 0$, $\theta \in (0, 1)$ and $\theta > 1$:

```python
plot_pairs(θ=-1, grid_max=1)
```

In the next case $\theta \in (0, 1)$ and stability is when $r(K) \leq 1$, because then $A$ is a contraction.  Note that it's not concave.

```python
plot_pairs(θ=0.2, grid_max=6)
```

When $\theta > 0$ we go back to the monotone concave case, but now $r(K) < 1$ is needed to cross the fixed point:

```python
plot_pairs(θ=1.4, grid_max=10)
```

The alternative is just to plot $\phi$:

Here's $\theta < 0$:

```python
g = np.linspace(1e-8, 2, 200)
fig, ax = plt.subplots(figsize=(6, 5))

ax.plot(g, ϕ(g, -1), label="$\phi$")  
ax.plot(g, g, 'k--', label="$45^\circ$")
ax.legend()

plt.show()
```

Here's $\theta \in (0, 1)$:

```python
g = np.linspace(1e-8, 2.5, 200)
fig, ax = plt.subplots(figsize=(6, 5))

ax.plot(g, ϕ(g, 0.25), label="$\phi$")  
ax.plot(g, g, 'k--', label="$45^\circ$")
ax.legend()

plt.show()
```

```python
g = np.linspace(1e-8, 15.0, 200)
fig, ax = plt.subplots(figsize=(6, 5))

ax.plot(g, ϕ(g, 1.25), label="$\phi$")  
ax.plot(g, g, 'k--', label="$45^\circ$")

ax.legend()

plt.show()
```

As per request:

```python

def plot_funcs(θ=1.0, grid_max=2.0):
    
    fig, ax = plt.subplots(figsize=(6, 5))
    
    g = np.linspace(1e-8, grid_max, 200)
    ax.plot(g, g, 'k-', lw=0.5, label="$45^\circ$")

    labels = "$r(K) > 1$", "$r(K)=1", "$r(K) < 1$"
    lts = "-", "-.", "--"

    for (K, label, lt) in zip ((2.0, 1.0, 0.5), labels, lts):

        ax.plot(g, ϕ(K * g, θ), label=label, linestyle=lt)

    ax.legend()
    ax.set_xlabel("$g$", fontsize=14)
    ax.set_ylabel("$Ag$", fontsize=14)
    ax.set_ylim(0, grid_max)
    ax.set_xlim(0, grid_max)
    
    plt.show()
```

```python
plot_funcs(θ=-0.5, grid_max=1.0)
```

```python
plot_funcs(θ=0.5, grid_max=10.0)
```

```python
plot_funcs(θ=1.5, grid_max=10.0)
```
