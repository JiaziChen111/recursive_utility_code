## Compute the fixed point of T for the SSY model

```julia
include("src/ssy_solution.jl")
```

```julia
using PyPlot
plt = PyPlot
```

```julia
ez = EpsteinZinSSY()
ssy = SSYconsumption()
```

```julia
scm = compute_solution(ez, ssy, verbose=true);
```

```julia
x_vals = scm.z_grid
y_vals = scm.h_c_grid
Z = scm.w_star[:, 2, :]
```

```julia
fig, ax = plt.subplots(figsize=(10, 5.7))

cs1 = ax[:contourf](x_vals, 
                    y_vals, 
                    Z', # cmap=plt.cm[:jet],
                    alpha=0.5)

ctr1 = ax[:contour](x_vals, 
                    y_vals, 
                    Z')

plt.clabel(ctr1, inline=1, fontsize=13)
plt.colorbar(cs1, ax=ax)

plt.show()
```
