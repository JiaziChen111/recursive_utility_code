```julia
using PyPlot
plt = PyPlot
```

```julia
include("src/stability_tests.jl")
```

```julia
K, I, J = 10, 10, 10
ez = EpsteinZinSSY()
ssy = SSYconsumption()
ssyd = discretize(ssy, K, I, J);
```

```julia
compute_spec_rad(ez, ssyd)
```

```julia
M = I * J * K
```

```julia
Km = compute_K(ez, ssyd);
```

```julia
r_vals = []
n = 2000
B = Km
w = ones(M) 
for i in 1:n
    r = maximum(abs, B * w)^(1/i)
    push!(r_vals, r)
    B = B * Km
end
```

```julia
fig, ax = plt.subplots()
ax[:plot](r_vals)
ax[:grid]()
plt.show()
```
